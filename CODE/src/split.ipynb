```python
import pandas as pd

path_economy = '..\data\Global Economy Indicators.csv'
df_economy = pd.read_csv(path_economy)
```

```python
# 删除国家ID列以及 Changes in inventories 还有Currenc、 AMA
df_economy = df_economy.drop(columns=[' CountryID '])
df_economy = df_economy.drop(columns=[' Changes in inventories '])
df_economy = df_economy.drop(columns=[' Currency '])
df_economy = df_economy.drop(columns=[' AMA exchange rate '])
df_economy = df_economy.drop(columns=[' IMF based exchange rate '])
df_economy.columns = df_economy.columns.str.strip()
```

```python
import os
ban_name = [" Sint Maarten (Dutch part) "," Czechoslovakia (Former) "," USSR (Former) ", " China, Macao SAR ",
            " D.P.R. of Korea "," Monaco ", " China, Hong Kong SAR "]
country_dfs = {country_name: group for country_name, group in df_economy.groupby("Country") if country_name not in ban_name}

output_path_all = '../data/all_countries_combined.csv'
output_dir_per_country = '../data/countries_data/'
os.makedirs(output_dir_per_country, exist_ok=True)
# 初始化列表用于后续合并
combined_dfs = []

for k,v in country_dfs.items():
    # 处理缺失值
    if v.isnull().values.any():
        v = v.fillna(v.mean(numeric_only=True))
    if v.isnull().values.any():
        print(f"{k}: 仍然存在缺失值！,表长：{len(v)}")
        print(v.isnull().sum())
        print("------------------")
    
    name = k.replace('.', '').replace(' ', '_').replace(':', "")
    
    file_path = os.path.join(output_dir_per_country, f"{name}.csv")
    if "Tanzania" in name:
        print(file_path)
    v.to_csv(file_path, index=False)
    
    combined_dfs.append(v)
    
# 合并为一个大表
df_combined = pd.concat(combined_dfs, ignore_index=True)

# 保存大表
df_combined.to_csv(output_path_all, index=False)
```
